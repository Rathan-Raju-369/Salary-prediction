```python
#divide the data into input and output
#Experience-Input x value 2-dimensional
#Salary- output y value 1-dimensional
x=df.iloc[0:30,0:1].values
x
```

```python
y=df.iloc[:,1].values
y
```

```python
# for small data set train and test variable is not required
# normalization also not required
```

```python
#run a classifier
from sklearn.linear_model import LinearRegression
model=LinearRegression()
```

```python
#fit the model
model.fit(x,y)
```

```python
#Predicting the output
y_pred=model.predict(x)
y_pred
```
